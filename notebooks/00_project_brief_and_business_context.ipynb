# 00 - Project Brief and Business Context

## Canadian Retail Credit Risk Analytics: Explainable Default Prediction, Portfolio Monitoring & Model Governance

## Business problem

A Canadian retail lender needs an early-warning view of borrower default risk. The business problem is not only to predict default, but to support responsible and explainable risk decisions that can be understood by credit strategy teams, portfolio managers, model validators, and non-technical stakeholders.

This notebook defines the business framing, target outcome, stakeholder questions, and governance boundaries before any model is trained.

## Why this matters for finance roles in Canada

This project is designed to show practical skills required in Canadian risk analytics roles:

- SQL/Python-style data preparation and analytical thinking
- Retail credit risk fundamentals: delinquency, exposure, portfolio segmentation, loss trade-offs
- Model development with clear baselines and challenger models
- Explainability and model governance documentation
- Stakeholder communication through business metrics, not only accuracy

## Initial modelling objective

**Objective:** Predict whether a borrower is likely to default, using borrower, loan, employment, and historical credit behaviour fields.

**Positive class:** `Defaulter = 1`

**Business interpretation:** A high-risk prediction should trigger review, risk-based pricing consideration, credit line management, collections prioritization, or additional verification — not an automatic adverse decision in this portfolio project.

## Stakeholder questions

| Stakeholder | Question this project should answer | Project output |
|---|---|---|
| Credit Risk Analyst | Which borrower segments have elevated default risk? | Portfolio monitoring notebook |
| Data Analyst | Are data quality, missingness, duplicates, and schema issues controlled? | Data quality reports |
| Model Developer | Which model performs best and why? | Model comparison table |
| Model Risk / Governance | Is the model explainable, stable, and documented? | Model card and validation summary |
| Business Stakeholder | What is the trade-off between catching defaults and over-reviewing good borrowers? | Threshold and cost analysis |

## Success metrics

Because default is normally an imbalanced event, this project will not optimize for accuracy alone.

Primary model-selection metric: **PR-AUC**

Secondary metrics:

- ROC-AUC
- Recall for defaulters
- Precision among reviewed/high-risk borrowers
- F1 score
- Balanced accuracy
- MCC
- Brier score / calibration
- Business-cost curve

## Governance boundaries

The project will explicitly review:

- Data leakage from repayment variables
- Protected or sensitive attributes and possible proxies
- Explainability for global and individual predictions
- Threshold choice based on business cost, not arbitrary 0.50 cutoff
- Model monitoring needs: drift, stability, and performance decay

This is a portfolio project and not a production lending system.
